# Results for Paper
Create charts and export tables for the IEEE paper and poster.

```
import pandas as pd
import matplotlib.pyplot as plt

metrics = pd.read_csv('../data/processed/aggregate_metrics.csv')
ax = metrics.set_index('model_name')[['rouge1_f1', 'rouge2_f1', 'rougeL_f1']].plot(kind='bar', figsize=(10, 6))
ax.set_title('ROUGE Comparison Across Models')
ax.set_ylabel('F1 Score')
plt.tight_layout()
plt.show()
```
